### 2.college 데이터 추출

```python
# 2.1 college.csv에서 INSTNM을 인덱스로 하는 college DataFrame을 생성하라

college = pd.read_csv('data/college.csv', index_col = 'INSTNM')
college.head()
```

```python
# 2.2 학부생 인원수가 4000명 이상인 학교 수를 구하라. 이는 전체 학교 중 몇 %에 해당하는 수치인가. 
```

```python
college['UGDS'].dropna(inplace = True)
```

```python
college['UGDS']
```

```python
# 4000명 이상 학교 수
(college['UGDS']>=4000).sum()
```

```python
# 4000명 이상 학교 비율
(college['UGDS']>=4000).sum()/college['UGDS'].count()
```

```python
# 2.3 학부생 인원이 500명 이하이면서 여학교 중 SAT 수학 점수 상위 10개 학교 목록을 작성하라
```

```python
c1 = college['UGDS']<= 500
c2 = college['WOMENONLY'] == 1
college[c1&c2].sort_values(by ='SATMTMID', ascending = False)[['UGDS','WOMENONLY','SATMTMID']]
```

```python
# 2.4 플로리다 주에 속하는 학교 중 ＇A＇~ ＇F＇로 시작하는 학교의 개수를 구하라
```

```python
college_sorted = college.sort_index()
```

```python
college_sorted[college_sorted['STABBR']=='FL'].loc['A':'G'].shape[0]
```

```python
# 2.5 상기 2번 대상 학교에 대해서 학부생 인원의 분포를 히스토그램으로 작성하라.
# 이 때, 크기를 (6,4)로 하고, 제목을 X축, Y축 명을 표기하라
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# stateful plt functions
plt.figure(figsize = (6,4))
plt.hist(college_sorted[college_sorted['STABBR']=='FL'].loc['A':'G']['UGDS'])
plt.ylabel('University Count')
plt.xlabel('UGDS')
plt.title('Student number distribution in Florida start with A~F')
plt.grid(True);
```

```python
# stateless
fig, ax = plt.subplots(figsize = (6,4))
ax.hist(college_sorted[college_sorted['STABBR']=='FL'].loc['A':'G']['UGDS'])
ax.set_ylabel('University Count')
ax.set_xlabel('UGDS')
ax.set_title('Student number distribution in Florida start with A~F')
ax.grid(True);
```
