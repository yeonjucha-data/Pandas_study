```python
import matplotlib.pyplot as plt
%matplotlib inline

# stateful plt functions
plt.figure(figsize = (6,4))
plt.hist(college_sorted[college_sorted['STABBR']=='FL'].loc['A':'G']['UGDS'])
plt.ylabel('University Count')
plt.xlabel('UGDS')
plt.title('Student number distribution in Florida start with A~F')
plt.grid(True);
```

```python
# stateless
fig, ax = plt.subplots(figsize = (6,4))
ax.hist(college_sorted[college_sorted['STABBR']=='FL'].loc['A':'G']['UGDS'])
ax.set_ylabel('University Count')
ax.set_xlabel('UGDS')
ax.set_title('Student number distribution in Florida start with A~F')
ax.grid(True);
```
